```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.ensemble import IsolationForest
df=pd.read_csv("embedded_system_network_security_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
features=df.drop(columns=['label'])
features.head()
```

```python
for col in df.columns:
    if df[col].dtype=='bool':
        df[col]=df[col].astype(int)
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
scaled_features_df=pd.DataFrame(scaled_features,columns=features.columns)
scaled_features_df.head()
```

```python
#parameters for isolation forest 
n_estimators=100  # number of trees in the forest
contamination=0.01 # proportion of outliers in the data
sample_size=256 # number of samples to draw from the data to train each tree
```

```python
#train the isolation forest model
model=IsolationForest(n_estimators=n_estimators,contamination=contamination,max_samples=sample_size,random_state=42)
model.fit(features)
#predict anomalies
anomaly_labels=model.predict(features)
#-1 for anomalies,1 for normal instances
```

```python
scaled_features_df['anomaly']=anomaly_labels
scaled_features_df.head()
```

```python
scaled_features_df['anomaly'].value_counts()
```

```python
# Visualize anomalies - Multiple feature comparisons
fig, axes = plt.subplots(1,2, figsize=(14, 5))
 
normal = scaled_features_df[scaled_features_df['anomaly'] == 1]
anomaly = scaled_features_df[scaled_features_df['anomaly'] == -1]
 
# Plot 1: CO vs C6H6
axes[0].scatter(normal['packet_size'], normal['inter_arrival_time'], color='blue', label='Normal', alpha=0.6, s=30)
axes[0].scatter(anomaly['packet_size'], anomaly['inter_arrival_time'], color='red', label='Anomaly', alpha=0.8, s=50, marker='x')
axes[0].set_xlabel('packet_size')
axes[0].set_ylabel('inter_arrival_time')
axes[0].set_title('packet_size vs inter_arrival_time')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
 
# Plot 2: NOx vs NO2
axes[1].scatter(normal['spectral_entropy'], normal['dst_port'], color='blue', label='Normal', alpha=0.6, s=30)
axes[1].scatter(anomaly['spectral_entropy'], anomaly['dst_port'], color='red', label='Anomaly', alpha=0.8, s=50, marker='x')
axes[1].set_xlabel('spectral_entropy')
axes[1].set_ylabel('dst_port')
axes[1].set_title('spectral_entropy vs dst_port')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
 
plt.tight_layout()
plt.show()
 
print(f"Normal data points: {len(normal)}")
print(f"Anomalies detected: {len(anomaly)}")
 
```
